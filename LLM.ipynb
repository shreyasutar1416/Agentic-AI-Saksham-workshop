#Initialize the LLM

```
from langchain_groq import ChatGroq

llm = ChatGroq(
    model = "openai/gpt-oss-120b",
    api_key = GROQ_API_KEY,
    temperature = 0.7,
    max_tokens = 256
)
```

#Run the LLM

```
response = llm.invoke("What is NLP AI?")
print(response)
```

```
response = llm.invoke("how to apply for scholarship?")
print(response)
```
